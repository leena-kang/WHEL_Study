```python
endpoints_df = pd.read_excel("C:/Users/leena/Niema_ML/data/endpoints.xls")
endpoints_df.head()
```

### subsetting dead and alive

```python
dead = endpoints_df[endpoints_df['vitality'] == 0].get(['id', 'vitality'])
alive = endpoints_df[endpoints_df['vitality'] == 1].get(['id', 'vitality'])
```

### blood carotenoids

```python
dead_blood = dead.merge(base_bloodcaro, left_on='id', right_on='ID').drop(columns=['ID'])
alive_blood = alive.merge(base_bloodcaro, left_on='id', right_on='ID').drop(columns=['ID'])
```

```python
alive_blood = alive_blood[alive_blood['betacar'] < 7.5]
alive_blood = alive_blood[alive_blood['alphacar'] < 2]
alive_blood = alive_blood[alive_blood['lutein'] < 2]
alive_blood = alive_blood[alive_blood['cryptox'] < 1]
```

#### visualizing distribution for each column

```python
cols = ['lutein', 'cryptox', 'lycopene', 'alphacar','betacar']
```

```python
sns.set_style('darkgrid')
i = 1
plt.figure(figsize=(15,15))
for col in cols:
    plt.subplot(5,2,i)
    sns.histplot(data=alive_blood, x=col, stat='density')
    plt.title('alive: ' + col)
    i += 1
    
    plt.subplot(5,2,i)
    sns.histplot(data=dead_blood, x=col,stat='density')
    plt.title('dead: ' + col)
    i += 1
```

### personal habits

```python
base_habits.head()
```

```python
# cleaning 'unknown' data 
habits_clean = base_habits[base_habits['smoking status'] != -9]
habits_clean = habits_clean[habits_clean['pack years'] != -9]
habits_clean = habits_clean[habits_clean['NEW METS'] != -9]
habits_clean = habits_clean[habits_clean['wgt history'] != -9]
```

```python
dead_habit = dead.merge(habits_clean, left_on='id', right_on='ID').drop(columns=['ID'])
alive_habit = alive.merge(habits_clean, left_on='id', right_on='ID').drop(columns=['ID'])
```

#### smoking status

1 - current smoker  
2 - former smoker  
3 - never smoker 

```python
def dead_freq(count):
    return count/dead_habit.shape[0]

def alive_freq(count): 
    return count/alive_habit.shape[0]

dead_habitcount = dead_habit.groupby('smoking status').count().reset_index()
dead_habitcount = dead_habitcount.assign(proportion=dead_habitcount['id'].apply(dead_freq))\
.get(['smoking status', 'id', 'proportion'])

alive_habitcount = alive_habit.groupby('smoking status').count().reset_index()
alive_habitcount = alive_habitcount.assign(proportion=alive_habitcount['id'].apply(alive_freq))\
.get(['smoking status', 'id', 'proportion'])
```

```python
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
ax = sns.barplot(x='smoking status', y='proportion', color='#6395d6', data=alive_habitcount)
ax.bar_label(ax.containers[0])
plt.title('Alive: Smoking Status')

plt.subplot(1,2,2)
ax = sns.barplot(x='smoking status', y='proportion', color='#6395d6', data=dead_habitcount)
ax.bar_label(ax.containers[0])
plt.title('Dead: Smoking Status')

plt.show()
```

#### NEW METS 

exercise METS/week  (Metabolic equivalents per week or kcal/week expended in mild, moderate, strenuous exercise and walking.)     

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
sns.histplot(data=alive_habit[alive_habit['NEW METS'] < 5000], x='NEW METS')
plt.title('Alive: NEW METS')

plt.subplot(1,2,2)
sns.histplot(data=dead_habit, x='NEW METS')
plt.title('Dead: NEW METS')

plt.show()
```
